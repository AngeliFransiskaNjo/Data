```python
from PIL import Image

# Buka gambar menggunakan PIL
photo = Image.open('mix.jpg') # gantilah 'mix.jpg' dengan nama file gambar Anda
photo = photo.convert('RGB')

# Ambil lebar (width) dan tinggi (height) gambar
lebar = photo.size[0]
tinggi = photo.size[1]

# Iterasi setiap baris (y) dan kolom (x) gambar
for y in range(0, tinggi):
    baris = ""
    for x in range(0, lebar):

        # Dapatkan nilai RGB dari piksel pada koordinat (x, y)
        RGB = photo.getpixel((x, y))
        R, G, B = RGB  # Gunakan nilai RGB tersebut

        # Tampilkan nilai R (Merah) ke konsol
        print(G)
        
import matplotlib.pyplot as plt
# Membuat histogram
plt.hist(R, bins=5, color='blue', edgecolor='black')
plt.xlim(1, 255)

# Menampilkan histogram
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data contoh
x = np.arange(1, 256)
y = np.sqrt(x)  # Contoh fungsi, sesuaikan dengan data Anda

# Plot data
plt.plot(x, y, label='Data')

# Menentukan skala sumbu X dan Y
plt.xscale('log')  # Skala logaritmik untuk sumbu X
plt.yscale('log')  # Skala logaritmik untuk sumbu Y

# Menambahkan judul dan label sumbu
plt.title('Contoh Plot dengan Skala Logaritmik (0-255)')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

# Menampilkan legenda
plt.legend()

# Menampilkan plot
plt.show()
```
